```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/telecom.csv")
df.head()
```

```python
selected_features = ['Int_Mem', 'Bty_Pwr', 'RAM']
df[selected_features].info()
df[selected_features].describe()
```

```python
int_mem = df['Int_Mem'].dropna()
mean_int_mem = int_mem.mean()
median_int_mem = int_mem.median()

plt.figure(figsize=(8, 5))
plt.hist(int_mem, bins=20, color='skyblue', edgecolor='black')
plt.axvline(mean_int_mem, color='red', linestyle='dashed', label=f'Mean: {mean_int_mem:.2f}')
plt.axvline(median_int_mem, color='green', linestyle='dotted', label=f'Median: {median_int_mem:.2f}')
plt.title('Histogram of Internal Memory')
plt.xlabel('Internal Memory')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
bty_pwr = df['Bty_Pwr'].dropna()
mean_bty_pwr = bty_pwr.mean()
median_bty_pwr = bty_pwr.median()

plt.figure(figsize=(8, 5))
plt.hist(bty_pwr, bins=20, color='lightcoral', edgecolor='black')
plt.axvline(mean_bty_pwr, color='blue', linestyle='dashed', label=f'Mean: {mean_bty_pwr:.2f}')
plt.axvline(median_bty_pwr, color='orange', linestyle='dotted', label=f'Median: {median_bty_pwr:.2f}')
plt.title('Histogram of Battery Power')
plt.xlabel('Battery Power')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
ram = df['RAM'].dropna()
mean_ram = ram.mean()
median_ram = ram.median()

plt.figure(figsize=(8, 5))
plt.hist(ram, bins=20, color='mediumseagreen', edgecolor='black')
plt.axvline(mean_ram, color='purple', linestyle='dashed', label=f'Mean: {mean_ram:.2f}')
plt.axvline(median_ram, color='gold', linestyle='dotted', label=f'Median: {median_ram:.2f}')
plt.title('Histogram of RAM')
plt.xlabel('RAM (MB)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['con5'] = (df['Int_Mem'] >= mean_int_mem) & (df['Bty_Pwr'] >= mean_bty_pwr) & (df['RAM'] >= mean_ram)
df[['Int_Mem', 'Bty_Pwr', 'RAM', 'con5']].head()
```

```python
df[['Depth', 'Weight']].info()
df[['Depth', 'Weight']].describe()
```

```python
depth = df['Depth'].dropna()
mean_depth = depth.mean()
median_depth = depth.median()

plt.figure(figsize=(8, 5))
plt.hist(depth, bins=20, color='cornflowerblue', edgecolor='black')
plt.axvline(mean_depth, color='red', linestyle='dashed', label=f'Mean: {mean_depth:.2f}')
plt.axvline(median_depth, color='green', linestyle='dotted', label=f'Median: {median_depth:.2f}')
plt.title('Histogram of Depth')
plt.xlabel('Depth (mm)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
weight = df['Weight'].dropna()
mean_weight = weight.mean()
median_weight = weight.median()

plt.figure(figsize=(8, 5))
plt.hist(weight, bins=20, color='orchid', edgecolor='black')
plt.axvline(mean_weight, color='navy', linestyle='dashed', label=f'Mean: {mean_weight:.2f}')
plt.axvline(median_weight, color='orange', linestyle='dotted', label=f'Median: {median_weight:.2f}')
plt.title('Histogram of Weight')
plt.xlabel('Weight (g)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['con6'] = (df['Depth'] <= median_depth) & (df['Weight'] <= median_weight)
df[['Depth', 'Weight', 'con6']].head()
```

```python
df1 = df[df['con5'] & df['con6']].copy()
df1.head()
```

```python
df1.shape
```

```python
df1_sorted = df1.sort_values(by='Price', ascending=True)
df1_sorted
```

```python
numeric_df = df1.select_dtypes(include='number')
deviations = numeric_df.std() / numeric_df.mean()
deviations.name = 'deviations'
deviations
```

```python
deviations_sorted = deviations.sort_values(ascending=False)
deviations_sorted
```
